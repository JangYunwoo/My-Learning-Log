### 1. DataFrame 생성

```python
import pandas as pd

data1 = {
    '이름': ['홍길동', '김철수', '이영희'],
    '나이': [25, 30, 22],
    '성별': ['남', '남', '여']
}
df1 = pd.DataFrame(data1)
print(df1)
```

```python
data2 = {
    '국어': [90, 85, 80],
    '수학': [95, 89, 76]
}
df2 = pd.DataFrame(data2, index=data1['이름'])
print(df2)
```

```python
data3 = {
    0: {'이름': '홍길동', '나이': 25},
    1: {'이름': '김철수', '나이': 30}
}
df3 = pd.DataFrame(data3).T
print(df3)
```

```python
data4 = [
    ['홍길동', 25, '남'],
    ['김철수', 30, '남'],
    ['이영희', 22, '여']
]
df4 = pd.DataFrame(data4, columns=['이름', '나이', '성별'])
print(df4)
```

```python
data5 = [
    ('홍길동', 25, '남'),
    ('김철수', 30, '남')
]
df5 = pd.DataFrame(data5, columns=['이름', '나이', '성별'])
print(df5)
```

```python
data6 = [
    {'이름': '홍길동', '나이': 25},
    {'이름': '김철수', '나이': 30}
]
df6 = pd.DataFrame(data6)
print(df6)
```

```python
import numpy as np

data7 = np.array([
    [1, 2],
    [3, 4],
    [5, 6]
])
df7 = pd.DataFrame(data7, columns=['A', 'B'])
print(df7)
```

```python
import json
json_str = '''
[
    {"이름": "홍길동", "나이": 25},
    {"이름": "김철수", "나이": 30}
]
'''
data8 = json.loads(json_str)
df8 = pd.DataFrame(data8)
print(df8)
```

### 3. DataFrame 중 일부 요소 선택

```python
df10 = pd.DataFrame({
    '이름': ['홍길동', '김철수', '이영희', '박영수', '최미나', '이민호'],
    '나이': [25, 30, 22, 28, 27, 30],
    '지역': ['서울', '부산', '서울', '대전', '서울', '부산'],
    '점수': [80, 90, 70, 85, 88, 91]
}, index=[101, 102, 103, 104, 105, 106])

print(df10[(df10['나이'] >= 27) & (df10['지역'] == '서울')])
```

```python
print(df10[(df10['지역'].isin(['서울', '부산']))])
```

```python
print(df10[(df10['지역'].isin(['서울', '부산'])) & (df10['점수'] > 85)])
```

```python
print(df10[df10.index.isin([102, 104, 106]) & (df10['점수'] > 88)])
```

```python
df11 = pd.DataFrame({
    '이름': ['홍길동', '김철수', '이영희', '박영수', '최미나'],
    '나이': [25, 30, 22, 28, 27],
    '지역': ['서울', '부산', '서울', '대전', '서울'],
    '점수': [80, 90, 70, 85, 88],
    '등록일': pd.to_datetime(['2023-01-01', '2022-12-15', '2023-02-10', '2023-01-20', '2023-03-05'])
})
cols = ['이름', '점수', '등록일']
df11_selected = df11[cols]
print(df11_selected)
```

```python
df11_filtered = df11.filter(like='일') # 이름에 '일'이 포함된 열만 선택
print(df11_filtered)
```

```python
df11_num = df11.select_dtypes(include=['int64']) # 필드가 숫자형(int64)인 열만 골라내기
print(df11_num)
```

```python
cols_cond = [col for col in df11.columns if '나' in col] # 이름에 '나'가 포함된 열만 선택
df11_cond = df11[cols_cond]
print(df11_cond)
```

```python
df12 = pd.DataFrame({
    '이름': ['홍길동', '김철수', '이영희', '박영수', '최미나'],
    '나이': [25, 30, 22, 28, 27],
    '지역': ['서울', '부산', '서울', '대전', '서울'],
    '점수': [80, 90, 70, 85, 88]
}, index=[101, 102, 103, 104, 105])
print(df12)
```

```python
print(df12.iloc[0, 1]) # 필드에서 좌표가 [0][1](index=0, col=1)인 데이터 추출
```

```python
print(df12.iloc[0:3, 1:3]) # x좌표가(index)가 0부터 2까지, y좌표(col)가 1부터 2까지인 데이터 추출
```

```python
print(df12.iloc[[1,2,4], [0,2]]) #index가 1, 2, 4이면서 col이 0, 2인 테이블 구성
```

```python
print(df12.loc[102, '점수']) # index, col을 이름으로 호출하며 그 좌표에 맞는 데이터 추출, 범위값 선택
```

```python
print(df12.loc[102:104, ['나이', '지역']]) # index가 102~104(슬라이싱과 범위 다름)이면서 col이 나이, 지역인 데이터
```

```python
print(df12.iat[3, 2]) # 단일값 선택용
```

```python
print(df12.at[104, '지역']) #단일값 선택용, index, col 이름으로 조회
```

```python
print(df12.loc[df12['나이'] >= 25, ['점수', '지역']]) #나이가 25이상인 것의 점수, 지역 col만 추출
```

```python
cond = (df12.index >= 102) & (df12['지역']).isin(['서울', '부산']) # index가 102이상이면서 지역이 서울,부산인 것의
print(df12.loc[cond, ['이름', '점수']]) # 이름, 점수행 출력
```
