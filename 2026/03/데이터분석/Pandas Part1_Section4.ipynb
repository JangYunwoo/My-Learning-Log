### 4. Dataframe 모양 변형 (reshape)

```python
import pandas as pd

df1 = pd.DataFrame({
    '날짜': ['2023-01-01', '2023-01-02'],
    '상품 A': [10, 15],
    '상품 B': [20, 15]
})

print(pd.melt(df1, id_vars=['날짜'], value_vars=['상품 A', '상품 B'], var_name='상품', value_name='판매량'))
```

```python
df1_long = pd.DataFrame({
    '날짜': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02'],
    '상품': ['상품 A', '상품 B', '상품 A', '상품 B'],
    '판매량': [10, 20, 15, 25]
})

print(df1_long.pivot(index='날짜', columns='상품', values='판매량'))
```

```python
df2 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df3 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

# 행 방향 연결
print(pd.concat([df2, df3], axis=0, ignore_index=True))
```

```python
# 열 방향 연결
print(pd.concat([df2, df3], axis=1))
```

### 5. Dataframe sorting, 칼럼/index 재설정

```python
df4 = pd.DataFrame({
    '부서': ['영업', '영업', '인사', '인사', '개발', '개발'],
    '이름': ['김철수', '이영희', '박영수', '최은지', '홍길동', '정해인'],
    '성과': [80, 70, 95, 90, 100, 85],
    '연봉': [5000, 4800, 5200, 5100, 6000, 5900] 
})

# 이름을 index로 설정
df4.set_index('이름', inplace=True)
print(df4)
```

```python
# 성과 내림차순, 연봉 오름차순 정렬
sorted_df4 = df4.sort_values(by=['성과', '연봉'], ascending=[False, True])
print(sorted_df4)
```

```python
# 인덱스를 열로 빼고 다시 기본 인덱스로 리셋
sorted_df4 = sorted_df4.reset_index()
print(sorted_df4)
```

```python
# 부서를 인덱스로 재설정
sorted_df4 = sorted_df4.set_index('부서')
print(sorted_df4)
```

```python
# 부서 순서대로 인덱스 정렬
sorted_df4 = sorted_df4.sort_index()
print(sorted_df4)
```

### 6. Dataframe 중복 값 / Null 값 탐지 및 제거

```python
data = {
    '이름': ['김철수', '이영희', '박민수', '이영희', '김철수', '김철수'],
    '부서': ['영업', '인사', '영업', '인사', '영업', '기획'],
    '연봉': [5000, 5200, 5000, 5200, 5000, 5500]
}

df5 = pd.DataFrame(data)
print(df5.duplicated(subset=['이름', '부서'], keep='last'))
```

```python
print(df5.drop_duplicates(subset=['이름', '부서'], keep='first'))
```

```python
# 연봉이 5000이상인 데이터에서 이름+부서 중복 제거
df5_filtered = df5[df5['연봉'] >= 5000].drop_duplicates(subset=['이름', '부서'])
print(df5_filtered)
```

```python
import numpy as np

df6  = pd.DataFrame({
    '이름': ['김철수', np.nan, '이영희', '박민수', None],
    '점수': [95, 85, np.nan, 75, None],
    '과목': ['수학', '영어', '수학', None, '국어'] 
})

# 결측치 여부 확인 - np.nan, None값 모두 True처리
print(df6.isnull())
```

```python
# 결측치가 있는 행만 보기, np.nan은 float형이므로 다른 int형 데이터도 float형으로 바뀜
print(df6[df6.isnull().any(axis=1)])
```

```python
# 이름이 결측치인 행만 선택
print(df6[df6['이름'].isnull()])
```

```python
# 전체 열에서 하나라도 NaN이 있으면 제거 - how의 기본값 = 'any'
print(df6.dropna())
```

```python
# 모든 열이 NaN인 행만 제거
print(df6.dropna(how='all'))
```

```python
# 특정 열만 고려하여 NaN 제거
print(df6.dropna(subset=['이름', '과목']))
```

```python
# NaN이 2개 이상인 행만 제거
print(df6.dropna(thresh=2))
```

```python
# 모든 NaN을 0으로 변경
print(df6.fillna(0))
```

```python
# 열마다 다른 값으로 채우기 이름-무명, 점수-평균, 과목-미정
print(df6.fillna({'이름': '무명', '점수': df6['점수'].mean(), '과목': '미정'}))
```

```python
# 앞의 값으로 채우기
print(df6.ffill())
```

```python
# 뒤의 값으로 채우기
print(df6.bfill())
```

```python
# 특정 열에서만 앞 값으로 채우기
df6['과목'] = df6['과목'].ffill()
print(df6)
```
